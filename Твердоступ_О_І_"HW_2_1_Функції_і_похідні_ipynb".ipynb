2. Використовуючи Python і `SymPy`, обчисліть похідні для функцій $ f(x) = \sin(x) $ та $ f(x) = \ln(x) $. Зобразіть ці функції та їхні похідні за допомогою `matplotlib`.



```python
from sympy import symbols, sin, ln, diff, lambdify

# Крок 1: Створюємо символічну змінну
x = symbols('x')

# Крок 2: Задаємо функції
f_sin = sin(x)
f_ln = ln(x)

# Крок 3: Обчислюємо похідні
f_sin_prime = diff(f_sin, x)    # похідна sin(x) = cos(x)
f_ln_prime = diff(f_ln, x)      # похідна ln(x) = 1/x

# Крок 4: Перетворення в числові функції
f_sin_func = lambdify(x, f_sin, 'numpy')
f_sin_prime_func = lambdify(x, f_sin_prime, 'numpy')

f_ln_func = lambdify(x, f_ln, 'numpy')
f_ln_prime_func = lambdify(x, f_ln_prime, 'numpy')

# Крок 5: Створюємо значення x для побудови графіків
x_sin_vals = np.linspace(-2 * np.pi, 2 * np.pi, 400)
x_ln_vals = np.linspace(0.1, 10, 400)  # починаємо з 0.1, бо ln(0) не існує

# Крок 6: Побудова графіків
plt.figure(figsize=(14, 6))

# Графік 1: sin(x) та похідна
plt.subplot(1, 2, 1)
plt.plot(x_sin_vals, f_sin_func(x_sin_vals), label="f(x) = sin(x)", color='blue')
plt.plot(x_sin_vals, f_sin_prime_func(x_sin_vals), label="f'(x) = cos(x)", color='red', linestyle='--')
plt.title("Функція та похідна: sin(x)")
plt.xlabel("x")
plt.ylabel("Значення")
plt.legend()
plt.grid(True)

# Графік 2: ln(x) та похідна
plt.subplot(1, 2, 2)
plt.plot(x_ln_vals, f_ln_func(x_ln_vals), label="f(x) = ln(x)", color='blue')
plt.plot(x_ln_vals, f_ln_prime_func(x_ln_vals), label="f'(x) = 1/x", color='red', linestyle='--')
plt.title("Функція та похідна: ln(x)")
plt.xlabel("x")
plt.ylabel("Значення")
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```


3. Загальні витрати компанії від обсягу виробництва задані функцією $ f(x) = x^2 - 5x + 1 $, де $ x $ — кількість вироблених одиниць. Обчисліть похідну цієї функції з допомогою SymPy та візуалізуйте значення функції і її похідної на відрізку від 0 до 100.

```python
x = symbols('x')

# Крок 2: функція загальних витрат
f = x**2 - 5*x + 1

# Крок 3: обчислення похідної
f_prime = diff(f, x)

# Крок 4: перетворення в числові функції для графіків
f_func = lambdify(x, f, 'numpy')
f_prime_func = lambdify(x, f_prime, 'numpy')

# Крок 5: створення масиву значень x від 0 до 100
x_vals = np.linspace(0, 100, 500)

# Обчислення значень функції та її похідної
y_vals = f_func(x_vals)
y_prime_vals = f_prime_func(x_vals)

# Крок 6: побудова графіків
plt.figure(figsize=(12, 6))

# Графік функції витрат
plt.plot(x_vals, y_vals, label="f(x) = x² - 5x + 1", color='blue')

# Графік похідної
plt.plot(x_vals, y_prime_vals, label="f'(x) = 2x - 5", color='red', linestyle='--')

# Оформлення
plt.title("Функція загальних витрат і її похідна")
plt.xlabel("Кількість вироблених одиниць (x)")
plt.ylabel("Значення")
plt.grid(True)
plt.legend()
plt.tight_layout()

# Показати графік
plt.show()
```
